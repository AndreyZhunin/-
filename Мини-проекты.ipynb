```python
from random import *    #Генератор паролей
digits = "0123456789"
lowercase_letters = "abcdefghijklmnopqrstuvwxyz"
uppercase_letters = "ABCDEFGHIJKLMNOPQRSTUVWXYZ"
punctuation = "!#$%&*+-=?@^_"
chars = ''
amount_passwords = int(input("Введите количество генерируемых паролей"))
len_password = int(input("Укажите длину пароля"))
numbers = input("Включать ли цифры в пароль?")
letters_upper = input("Включать ли прописные буквы?")
letters_lower = input("Включать ли строчные буквы?")
symbols = input('Включать ли символы "!#$%&*+-=?@^_"?')
ambiguous_symbols = input('Исключать ли неоднозначные символы "il1Lo0O"')
if numbers.lower() == "да" or numbers.lower() == "yes":
    chars += digits
if letters_upper.lower() == "да" or letters_upper.lower() == "yes":
    chars += uppercase_letters
if letters_lower.lower() == "да" or letters_lower.lower() == "yes":
    chars += lowercase_letters
if symbols.lower() == "да" or symbols.lower() == "yes":
    chars += punctuation
if ambiguous_symbols.lower() == "да" or ambiguous_symbols.lower() == "yes":
    for i in chars:
        if i in "il1Lo0O":
            chars = chars.remove(i)
def generate_password(length, chars):
    password = []
    for i in range(amount_passwords):
        password += sample(chars, length)
        password += " "
    password = "".join(password)
    password = password.split()
    return password
print(*generate_password(len_password, chars), sep="\n")
```

```python
from math import *    #Шифр Цезаря
encryption = input('Шифрование (+); Дешифрование(-)')
language = input('Русский (ру); Английский (en)')
step = int(input('Введите шаг сдвига'))
text = input("Введите текст")
if language.lower() == "ру" or language.lower() == "рус" or language.lower() == "ru" or language.lower() == "rus":
    power, power_min_low, power_min_upp = 32, 1072, 1040
elif language.lower() == "en" or language.lower() == "eng" or language.lower() == "англ":
    power, power_min_low, power_min_upp = 26, 97, 65
if encryption == "-":
    step *= (-1)
for i in text:
    if power_min_low <= ord(i) <= power_min_low + power - 1:
        z = (ord(i) + step - power_min_low) % power + power_min_low
        print(chr(z), end="")
    elif power_min_upp <= ord(i) <= power_min_upp + power - 1:
        z = (ord(i) + step - power_min_upp) % power + power_min_upp
        print(chr(z), end="")
    else:
        print(i, end="")
```
